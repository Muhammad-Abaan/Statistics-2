##### **3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution**

```python
def bernoulli_pdf(p):
    x = [1,0]
    probs = stats.bernoulli.pmf(x, p) 
    return dict(zip(x, probs))

pdf = bernoulli_pdf(0.7)
print("Bernoulli PDF:", pdf)
```

##### **9. Write a Python function to calculate Z-scores from a dataset and plot them**

```python
np.random.seed(1)
data = np.random.normal(50, 10, 100)

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return [(i-mean)/std for i in data]

data = np.random.normal(50, 10, 100)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, color='plum', edgecolor='black', density=True)
sns.kdeplot(z_scores, bw_adjust=1)
plt.title("Z-Score Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

##### **12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)**

```python
def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.plot(x, y, label='Standard Normal PDF', color='teal')
    plt.title("Standard Normal Distribution (mean=0, std=1)")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

plot_standard_normal()
```

##### **14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution**

```python
def calculate_z_score(x, mean, std):
    return (x - mean) / std

x = 75
mean = 70
std = 5

z = calculate_z_score(x, mean, std)
print(f"Z-score for x = {x} is {z:.2f}")


x_vals = np.linspace(-4, 4, 1000)
y_vals = stats.norm.pdf(x_vals)

plt.plot(x_vals, y_vals, color='green', label='Standard Normal')
plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
plt.title("Z-Score Visualization")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

##### **15. Implement hypothesis testing using Z-statistics for a sample dataset**

```python
population = np.random.normal(size=1000, loc=100, scale=2)
sample = np.random.choice(population, 200)

null_hypo = 100

zstat, pvalue = ztest(sample, value=null_hypo, alternative='larger')

if pvalue <= 0.05:
    print('Reject the null hypothesis')
else:
    print('Failed to reject the null hypothesis')
```

- The ztest() function assumes the sample size is large enough (n ≥ 30) for the Central Limit Theorem to hold, meaning the sample standard deviation approximates the population standard deviation. So even though population σ isn’t provided, it estimates it from the sample and uses the normal distribution for inference.

##### **16. Create a confidence interval for a dataset using Python and interpret the result**

```python
data = np.array([10, 12, 9, 11, 13, 12, 14, 10])
mean = np.mean(data)
sem = stats.sem(data)
confidence = 0.95
interval = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)

print(f"95% confidence interval: {interval}")
```

##### **17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean**

##### **23. Implement Z-score normalization on a dataset and visualize its transformation**

```python
data = np.random.randint(20, 80, 100)
z_scores = (data - np.mean(data)) / np.std(data)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(data, bins=10, color='orange', edgecolor='black')
plt.title("Original Data")

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=10, color='skyblue', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.tight_layout()
plt.show()
```
